Drop missing values

Note : If you're sure you want to drop rows with missing values, pandas does have a handy function, dropna()
***

```python
#Removes all the rows that contain a missing value
raw_data = pd.read_csv("../data/raw/wineratings.csv")
missing_values_count = raw_data.isnull().sum()
drop_missing_values = raw_data.dropna()
print(drop_missing_values)
```

Remove all the columns with at least one missing value


****


```python
#Remove all columns with at least one missing value 
raw_data = pd.read_csv("../data/raw/wineratings.csv")
columns_with_na_dropped = raw_data.dropna(axis=1)
columns_with_na_dropped.head()
```

#Just how much data did we lose ?

***

```python
#Just how much data did we lose ?
print("Columns in original dataset : %d \n" % raw_data.shape[1])

print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```
